# Wordnet Lemmatizer

Lemmitization technique is like stemming. The output we will get after lemmatization is called 'lemma', which is root word rather than the root stem, the output of stemming. After lemmatization we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thinwrapper around the wordnet corpus. This class use morphy() function to the WordNet CorpusReader class to find a lemma. 

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "historical", "finally", "finalized"]
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# must download
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word))
```

### Advantage of Lemmatization

Lemmatization converts words into their meaningful base (dictionary) form, ensuring better linguistic accuracy in NLP tasks.

### Disadvantage of Lemmatization

Lemmatization is computationally slower because it relies on language dictionaries and POS tagging.

### Use Case of Lemmatization

Lemmatization is widely used in sentiment analysis and text classification, where preserving the actual meaning of words is important.
