# Stop Words

Stop words are commonly used words in a language that usually do not add much meaningful information to the text analysis. Examples include words like “is”, “am”, “the”, “and”, “in”, “on”, etc.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# must download
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = 'Hi! My name is Purvi Jain. I am 18 years old'
```

```python
tokens = word_tokenize(text)
print(tokens)
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
filtered_words = []

for word in tokens:
    if word.lower() not in stop_words:
        filtered_words.append(word)

print(filtered_words)
```

```python
clean_text = " ".join(filtered_words)
print(clean_text)
```
